```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import scipy.stats as stats
import statsmodels.api as sm
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("Cutlets.csv")
```

```python
df
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
plt.boxplot(df['Unit A'])
plt.title("Unit A")
plt.subplot(1,2,2)
plt.boxplot(df['Unit B'])
plt.title("Unit B")
```

```python
plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
sns.histplot(df['Unit A'],bins=15)
plt.xlabel('Values in Unit A')
plt.ylabel('Frequency')
plt.title('Histogram of Unit A')
plt.subplot(1,2,2)
sns.histplot(df['Unit B'],bins=15)
plt.xlabel('Values in Unit B')
plt.ylabel('Frequency')
plt.title('Histogram of Unit B')
```

```python
fig,ax=plt.subplots(1,2,figsize=(10,3))
sm.qqplot(df["Unit A"], line = 'q',ax=ax[0])
plt.title('Unit A')
sm.qqplot(df["Unit B"], line = 'q',ax=ax[1])
plt.title('Unit B')
```

Compare Evidences with Hypothesis using t-statistics

```python
statistic , p_value = stats.ttest_ind(df['Unit A'],df['Unit B'], alternative = 'two-sided')
```

```python
print('p_values =', round(p_value,4))
```

```python
alpha = 0.05
print(f'Significance={alpha:.2f}',f'p={p_value:.4f}')
if p_value <= alpha:
    print('We reject Null Hypothesis there is a significance difference between two Units A and B')
else:
    print('We fail to reject Null hypothesis')
```

# Hence, We fail to reject Null Hypothesis because of lack of evidence, there is no significant difference between the two samples

```python
import numpy as np
from scipy.stats import chi2_contingency
```

```python
br=pd.read_csv('BuyerRatio.csv',index_col=0)
br
```

```python
observed_values=np.array([[50,142,131,70],[435,1523,1356,750]])
```

```python
chi2, p_value, _, _ = chi2_contingency(observed_values)
```

```python
print(f'Chi-Square Statistic: {chi2:.4f}')
print(f'P-value: {p_value:.4f}')
```

```python
alpha = 0.05
print(f'Significance={alpha:.2f}',f'p={p_value:.4f}')
if p_value <= alpha:
    print('We reject Null Hypothesis there is a significance difference between two Units A and B')
else:
    print('We fail to reject Null hypothesis')
```

```python
observed = np.array([271, 267, 269, 280, 29, 33, 31, 20])
expected = np.array([271.75, 271.75, 271.75, 271.75, 28.25, 28.25, 28.25, 28.25])
```

```python
statistic , p_value = stats.chisquare(observed, expected, ddof = df)
print("Statistic = ",statistic,'\n', 'p_value =',p_value)
```

```python
alpha = 0.05
print(f'Significance={alpha:.2f}',f'p={p_value:.4f}')
if p_value <= alpha:
    print('We reject Null Hypothesis there is a significance difference between two Units A and B')
else:
    print('We fail to reject Null hypothesis')
```
